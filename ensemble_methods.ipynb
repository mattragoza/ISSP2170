## Results

```python
jobs = pd.concat([
    prob1_jobs,
    prob2_jobs,
    prob3_jobs,
    prob4_jobs,
    prob5_jobs
])

jobs['n_trees'] = jobs['n_trees'].fillna(1).astype(int)

jobs.set_index(['problem', 'option', 'n_trees', 'job_name'])
```

```python
# load all experiment outputs

metrics = ps.metrics(jobs, metric_pat='(.+)csv', names=['data', 'y_true', 'y_pred', 'prob'])

index_cols = ['problem', 'option', 'n_trees', 'job_name', 'data']
output_cols = ['y_true', 'y_pred', 'prob']

metrics.groupby(index_cols)[output_cols].count()
```

```python
def evaluate_classifier(x):
    import sklearn.metrics
    ret = dict()
    ret['accuracy'] = sklearn.metrics.accuracy_score(x['y_true'], x['y_pred'])
    ret['AUROC'] = sklearn.metrics.roc_auc_score(x['y_true'], x['prob'])
    return pd.Series(ret)

# aggregate metrics

agg = metrics.groupby(index_cols).apply(evaluate_classifier).reset_index()
agg.set_index(index_cols).unstack().sort_index(axis=1, ascending=False)
```

```python
# problem 1 table
prob1_tab = agg[
    (agg['problem'] == 1)
].set_index(['job_name', 'data'])[['accuracy', 'AUROC']].unstack() \
    .sort_index(axis=1, ascending=False)

prob1_tab
```

```python
print(prob1_tab.to_latex(float_format='%.3f'))
```

```python
# problem 2 table
prob2_tab = agg[
    (agg['problem'].isin({1, 2}))
].set_index(index_cols[1:])[['accuracy', 'AUROC']].unstack() \
    .sort_index(axis=0, level=[0,1]).sort_index(axis=1, ascending=False)

prob2_tab
```

```python
print(prob2_tab.to_latex(float_format='%.3f'))
```
